# Desafio 3

Queremos conhecer melhor nossos clientes por estado. Para isso, iniciamos uma análise na pontuação de crédito. 
Para realizar a verificação inicial, precisamos de alguns valores.
Os valores são a média, a mediana, a moda e o desvio padrão da pontuação de crédito.

## _Set up_ 

```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv("desafio1.csv", index_col = 'estado_residencia')
```

## Inicie sua análise a partir daqui

```python
df
```

## Eliminação de colunas desnecessárias


```python
df.drop(df.columns.difference(['pontuacao_credito']), 1, inplace=True)
df
```

## Moda


```python
df['moda'] = df.groupby('estado_residencia').apply(lambda x: x.mode().iloc[0])
df
```

## Mediana


```python
df['mediana'] = df.groupby('estado_residencia')['pontuacao_credito'].median()
df
```

## Média


```python
df['media'] = df.groupby('estado_residencia')['pontuacao_credito'].mean()
df
```

## Desvio Padrão


```python
df['desvio_padrao'] = df.groupby('estado_residencia')['pontuacao_credito'].std()
df
```

## Exportando resultado


```python
df.drop('pontuacao_credito', axis=1, inplace=True)
df.drop_duplicates(inplace=True)
df.to_json('submission.json', orient='index')
df
```
